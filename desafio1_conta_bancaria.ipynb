```python
class Usuario:
    def __init__(self, nome, cpf, endereco, ano_nascimento, estado, email, numero_celular):
        self.nome = nome
        self.cpf = cpf
        self.endereco = endereco
        self.ano_nascimento = ano_nascimento
        self.estado = estado
        self.email = email
        self.numero_celular = numero_celular
        self.operacoes = []

class ContaCorrente:
    numero_banco = 495

    def __init__(self, numero_conta, saldo):
        self.numero_conta = numero_conta
        self.saldo = saldo

    def depositar(self, valor):
        self.saldo += valor
        self.registrar_operacao(f"Depósito: +R$ {valor:.2f}")

    def sacar(self, valor):
        if valor > self.saldo:
            print("Saldo insuficiente. Transação não realizada.")
        else:
            self.saldo -= valor
            self.registrar_operacao(f"Saque: -R$ {valor:.2f}")

    def registrar_operacao(self, operacao):
        self.operacoes.append(operacao)

    def visualizar_historico(self):
        print("Histórico de Operações:")
        for operacao in self.operacoes:
            print(operacao)

def criar_usuario():
    nome = input("Nome: ")
    cpf = input("CPF (11 dígitos): ")
    endereco = input("Endereço: ")
    ano_nascimento = input("Ano de Nascimento: ")
    estado = input("Estado: ")
    email = input("Email: ")
    numero_celular = input("Número de Celular: ")

    if len(cpf) != 11:
        print("CPF inválido. Por favor, insira um CPF com 11 dígitos.")
        return None

    return Usuario(nome, cpf, endereco, ano_nascimento, estado, email, numero_celular)

def criar_conta_corrente(usuario):
    numero_conta = str(len(usuario.cpf)).zfill(8)  # Número da conta com base no tamanho do CPF
    saldo = float(input("Saldo em Reais: "))

    return ContaCorrente(numero_conta, saldo)

# Exemplo de uso
usuario = criar_usuario()
conta_corrente = criar_conta_corrente(usuario)

print("Usuário criado com sucesso!")
print("Dados do Usuário:")
print(f"Nome: {usuario.nome}")
print(f"CPF: {usuario.cpf}")
print(f"Endereço: {usuario.endereco}")
print(f"Ano de Nascimento: {usuario.ano_nascimento}")
print(f"Estado: {usuario.estado}")
print(f"Email: {usuario.email}")
print(f"Número de Celular: {usuario.numero_celular}")

print("\nConta Corrente criada com sucesso!")
print("Dados da Conta Corrente:")
print(f"Número do Banco: {ContaCorrente.numero_banco}")
print(f"Número da Conta: {conta_corrente.numero_conta}")
print(f"Saldo: R$ {conta_corrente.saldo:.2f}")

conta_corrente.depositar(1000)
conta_corrente.sacar(500)
conta_corrente.sacar(2000)  # Tentativa de sacar um valor maior que o saldo disponível

conta_corrente.visualizar_historico()
```
